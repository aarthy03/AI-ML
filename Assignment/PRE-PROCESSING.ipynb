### OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct=ColumnTransformer([('encoder',OneHotEncoder(),[-1])],remainder='passthrough')
```

```python
ct
```

```python
ds=np.array(ct.fit_transform(ds),dtype=str)
```
